```python
# Starting the browser and openning Printify
global browser


def start_google_chrome():
    try:
        print("Please Wait. Launching Chrome")
        global browser
        options = webdriver.ChromeOptions()
        options.page_load_strategy = 'normal'
        prefs = {"profile.default_content_setting_values.notifications": 2,
                 "credentials_enable_service": False,
                 "profile.password_manager_enabled": False}
        options.add_experimental_option("prefs", prefs)
        # options.add_argument("start-maximized")
        options.add_argument("--disable-extensions")
        # options.add_argument("--headless")
        options.add_experimental_option("excludeSwitches", ["enable-automation"])
        options.add_experimental_option('useAutomationExtension', False)
        browser = webdriver.Chrome(options=options,
                                   executable_path=r'C:\Program Files\Google\Chrome\Application\chromedriver.exe')
        time.sleep(2)
        browser.set_window_size(1000, 1100)
        print("Browser launched successfully")

    except Exception as e:
        print("Could not start Chrome: ", e)
        # start_odibets()


def open_printify():
    # Open the first tab for product catalogue
    browser.switch_to.window(browser.window_handles[0])
    url = "https://printify.com/app/products"
    browser.get(url)
    

start_google_chrome()
time.sleep(4)
open_printify()
```

### Once the page is loaded, I manually use navigate to the page  I want to capture data and iteratively run the cell below to capture that rendered data.

```python
# This cell captures all the products rendered on a page after I manually 
# navigate to the site.
all_products_showing = browser.find_elements_by_class_name("product-card-container")

for product in all_products_showing:
    # Append each products details on to the respective variable above
    # after setting it manually
    home_living_products.append(product.text)
    # print(product.text)
print(len(home_living_products))  # Track the size of the category each time a capture is made
```

```python
# Attach the as value data into appropriate key
printify_products_dict["Home & Living"] = home_living_products
print(len(printify_products_dict["Home & Living"]))
print(printify_products_dict["Home & Living"])
```

```python
# Takes a list from dict value and append it to an excel file and save it
def save_to_excel(result):
    workbook_obj = openpyxl.load_workbook("printify_products.xlsx")
    sheet_obj = workbook_obj.active
    for details in result:
        details_lst = details.split('\n')
        sheet_obj.append(details_lst)
    workbook_obj.save("printify_products.xlsx")
```
